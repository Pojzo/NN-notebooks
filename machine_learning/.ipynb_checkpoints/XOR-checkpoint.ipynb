```python
x = [0, 1]
t = [1, 0]

w1 = np.random.randn(1)
b = np.random.randn(1)

def forward(x, w, b):
    z = w * x + b
    y = sigmoid(z)
    return y, z

def backward(x, t, y, z):
    derror_y = y - t
    dy_z = y * (1 - y)
    dz_w = x
    dz_b = np.ones_like(x)
    
    derror_w = np.mean(derror_y * dy_z * dz_w)
    derror_b = np.mean(derror_y * dy_z * dz_b)
    
    return derror_w, derror_b

lr = 0.5

for i in range(1000):
    y, z = forward(x, w1, b)

    error = np.mean(0.5 * (t - y) ** 2)
    
    derror_w1, derror_b = backward(x, t, y, z)
    
    w1 -= lr * derror_w1
    b -= lr * derror_b
    print(error)
```
