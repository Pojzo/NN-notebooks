```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def sin(x):
    return np.sin(x)

def cos(x):
    return np.cos(x)
```

```python
x = np.arange(-np.pi * 2, np.pi * 2, np.pi/1000)
y_sin = sin(x)
y_cos = cos(x)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

ax1.plot(x, y_sin, label='f(x) = sin(x)')
ax1.plot(x, np.zeros_like(x), color='gray', linestyle='dashed')
ax1.axvline(x = 0, color = 'gray', linestyle='dashed')
ax1.legend()

ax2.plot(x, y_cos, label='f(x) = cos(x)', color='green')
ax2.plot(x, np.zeros_like(x), color='gray', linestyle='dashed')
ax2.axvline(x = 0, color = 'gray', linestyle='dashed')
ax2.legend()

plt.tight_layout()
```

```python
import torch
import torch.nn.functional as F
```

```python
lr = 0.005
epochs = 1000
dense_neurons = 200
X = torch.tensor(x, dtype=torch.float32).reshape(len(x), 1)
Y_sin = torch.tensor(y_sin, dtype=torch.float32)
Y_cos = torch.tensor(y_cos, dtype=torch.float32)
W = torch.randn((1, dense_neurons), requires_grad=True)
W2 = torch.randn((dense_neurons, dense_neurons), requires_grad=True)

bias = torch.randn((1, dense_neurons), requires_grad=True)
bias2 = torch.randn((1, dense_neurons), requires_grad=True)

history = []

loss = 5

for i in range(epochs):
    layer1_output = X @ W + bias
    activation1 = torch.tanh(layer1_output)
    layer2_output = activation1 @ W2 + bias2
    activation_output = torch.tanh(layer2_output)
    s_output = activation_output.sum(axis=1)
    loss = F.mse_loss(s_output, Y_sin)
    W.grad = None
    bias.grad = None
    W2.grad = None
    bias2.grad = None
    loss.backward()
    W.data += -lr * W.grad
    W2.data += -lr * W2.grad
    bias.data += -lr * bias.grad
    bias2.data += -lr * bias2.grad
    print(f'loss = {loss.item()} epoch = {i}')
    
    
    history.append(loss.data.item())
```

```python
plt.plot(history)
plt.xlabel("Epochs")
plt.ylabel("Mean squared loss")
print(loss)
```
