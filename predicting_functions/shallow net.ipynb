```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
def sigmoid_d(x):
    return 1 - sigmoid(x)
```

```python
import numpy as np
x = np.arange(0, 100, 1).reshape(100, 1) / 100
y = x.copy()
w = np.random.randn()
b = np.random.randn()

def forward(w, x, b):
    return w * x + b

z = forward(w, x, b)
a = sigmoid(z)
loss = ((a - y) ** 2).sum()
plt.plot(x, y * 100, label = 'True value')
plt.plot(x, a * 100, label = 'Model prediction')
plt.legend()
print('output of the model with randomly initialized weight and bias')
```

```python
lr = 0.1
epochs = 10000

print(f'Running model for {epochs} epochs, with {lr=}')
for i in range(epochs):
    # propagate forward
    z = forward(w, x, b)
    a = sigmoid(z)
    loss = ((a - y) ** 2).mean()
    
    
    # calculate derivatives 
    dloss_da = 2 * (a - y)
    da_dz = sigmoid_d(z)
    dz_dw = x.copy()
    dz_db = 1
    dloss_dw = (dloss_da * da_dz * dz_dw).mean()
    dloss_db = (dloss_da * da_dz * dz_db).mean()

    # update weight and bias based on calculated partial derivatives
    w -= dloss_dw * lr
    b -= dloss_db * lr
print(f'\n{loss=}')
```

```python
plt.plot(x * 100, y * 100, label = 'True value')
plt.plot(x * 100, a * 100, label = 'Model prediction')
plt.legend()
print(w, b)
# model prediction after training
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='rmsprop', loss='mse')
model.fit(x, y, epochs=1000, verbose=1)
```

```python
plt.plot(x * 100, y * 100, label = 'True value')
plt.plot(x * 100, model.predict(x) * 100, label = 'Model prediction')
plt.legend()
```

```python
import numpy as np
x = np.arange(-5, 5 + 0.1, 0.1).reshape(101, 1)
y = x ** 2

# normalize the input data
x_max = x.max()
y_max = y.max()
x /= x_max
y /= y_max
w = np.random.randn()
b = np.random.randn()

lr = 0.1
epochs = 1000

print(f'Running model for {epochs} epochs, with {lr=}')
for i in range(epochs):
    # propagate forward
    z = forward(w, x, b)
    a = sigmoid(z)
    loss = ((a - y) ** 2).mean()
    
    # calculate derivatives 
    dloss_da = 2 * (a - y)
    da_dz = sigmoid_d(z)
    dz_dw = x.copy()
    dz_db = 1
    dloss_dw = (dloss_da * da_dz * dz_dw).mean()
    dloss_db = (dloss_da * da_dz * dz_db).mean()

    # update weight and bias based on calculated partial derivatives
    w -= dloss_dw * lr
    b -= dloss_db * lr
print(f'\n{loss=}')
```

```python
plt.plot(x * x_max, y * y_max, label = 'True value')
plt.plot(x * x_max, a * y_max, label = 'Model prediction')
plt.legend()
print(w, b)
```
